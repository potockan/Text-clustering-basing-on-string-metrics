```python
import scipy.ndimage
import numpy as np
y = scipy.ndimage.imread("/home/samba/potockan/python/351451.jpg")
y.shape
y = y/255
y = np.mean(y, axis = 2)
y.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.axis('off')
plt.imshow(y, cmap = plt.cm.gray)
```

```python
a = np.array([
        [-1,-1,-1],
        [-1,8.5,-1],
        [-1,-1,-1]
    ])
import scipy.signal
y2 = scipy.signal.convolve2d(y, a)
y2[y2>0.5] = 1
y2[y2<0.5] = 0
plt.axis('off')
plt.imshow(y2, cmap = plt.cm.gray)
```

```python
a = np.array([
        [1/9,1/9,1/9],
        [1/9,1/9,1/9],
        [1/9,1/9,1/9]
    ])
import scipy.signal
y2 = scipy.signal.convolve2d(y, a)
y2[y2>0.5] = 1
y2[y2<0.5] = 0
plt.axis('off')
plt.imshow(y2, cmap = plt.cm.gray)
```
