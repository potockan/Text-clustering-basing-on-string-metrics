Łączenie programów w łańcuch. Każdy program konsolowy (też pythonowe)
ma otwarte 3 urządzenia:
* stdin - domyślnie klawiatura
* stdout - domyślnie konsola
* stderr - domyślnie konsola
Te urządzenia mozna przekierować, np. `cat sfjo` + CTRL+D

Można przekierować stdout: `cat > plik.txt`
`cat >> plik.txt` - dopisuje
`cat < plik.txt` - stdin

pipe: `|`

## Skrypty pythona w konsoli

Tworzymy w aktualnym katalogu skrypt
`test.py`:

```python
print("aloha omm" +__name__)
print(2+2)
```

Zwrócmy uwagę na `__name__` - gdy skrypt uruchamiamy w kosoli
jest równe `__main__` a nie nazwa modulu

Można więc pisać moduło-skrypty

```python
def kwadrat(x):
    '''
    Liczy kwadrat danej liczby.

    >>> kwadrat(2)
    4
    >>> kwadrat(3)
    9
    >>> kwadrat(4) # to sie wywali!
    9
    '''
    return x ** 2

if __name__ == '__main__':
    #testy jednostkowe
    assert kwadrat(2) == 4
    assert kwadrat(3) == 9
    for i in range(0, 10):
        assert kwadrat(i) == i**2
    import doctest
    doctest.testmod()    
    # doctesty - troche automatyzacji
```
    

BTW na temat pakietow i scieżek wyszukiwania.

Domyślnie python szuka pakietów:
* w bieżącym katalogu
* w katalogach wyminienionych w zmiennej środowiskowej PYTHONPATH
* w miejscach określonych przez dystrybucję pythona

Można pisac skrypty uruchamialne tak jak 
kazdy inny linuxowy program (bez jawnego podawania interpretera)
Żeby uruchomić np. plik `test.py` trzeba dodać na początku 
`#! /opt/anaconda/bin/python3`
nadać uprawnienia `chmod 775 test.py`
i można wywołać `./test.py`
